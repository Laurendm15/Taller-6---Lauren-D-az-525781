```python

usuarios = [
    {'id': 1, 'nombre': 'Ren Tao', 'contraseña': '1234_4321', 'rol': 'admin'},
    {'id': 2, 'nombre': 'Ioh Asakura', 'contraseña': 'abc_cba', 'rol': 'vendedor'},
    {'id': 3, 'nombre': 'Chocolove McDonnell', 'contraseña': '11110000', 'rol': 'vendedor'}
]

productos = [
    {'id': 1, 'nombre': 'Ladrillo 10x20x30', 'precio_compra': 1200.00, 'precio_venta': 1500.00, 'cantidad': 1000, 'unidad': 'un'},
    {'id': 2, 'nombre': 'Cemento Gris x50kg', 'precio_compra': 28000.00, 'precio_venta': 30000.00, 'cantidad': 200, 'unidad': 'saco'},
    {'id': 3, 'nombre': 'Varilla Corrugada #5 x6m', 'precio_compra': 33000.00, 'precio_venta': 35000.00, 'cantidad': 150, 'unidad': 'un'},
    {'id': 4, 'nombre': 'Impermeabilizante 22kg', 'precio_compra': 350000.00, 'precio_venta': 360000.00, 'cantidad': 50, 'unidad': 'un'},
    {'id': 5, 'nombre': 'Malla Electro Soldada 6x2.2m', 'precio_compra': 42000.00, 'precio_venta': 45000.00, 'cantidad': 75, 'unidad': 'un'}
]

ventas = []  # Lista para almacenar las ventas

def iniciar_sesion(nombre, contraseña):
    for usuario in usuarios:
        if usuario['nombre'] == nombre and usuario['contraseña'] == contraseña:
            return usuario['rol']
    return None

def registrar_usuario(nombre, contraseña, rol):
    nuevo_id = len(usuarios) + 1
    usuarios.append({'id': nuevo_id, 'nombre': nombre, 'contraseña': contraseña, 'rol': rol})
    print(f'Usuario {nombre} registrado con éxito.')


def agregar_actualizar_producto(id_producto, nombre, precio_compra, precio_venta, cantidad, unidad):
    for producto in productos:
        if producto['id'] == id_producto:

            producto.update({'nombre': nombre, 'precio_compra': precio_compra, 'precio_venta': precio_venta, 'cantidad': cantidad, 'unidad': unidad})
            print(f'Producto {nombre} actualizado con éxito.')
            return

    productos.append({'id': id_producto, 'nombre': nombre, 'precio_compra': precio_compra, 'precio_venta': precio_venta, 'cantidad': cantidad, 'unidad': unidad})
    print(f'Producto {nombre} agregado con éxito.')


def consultar_inventario(codigo=None):
    if codigo:
        for producto in productos:
            if producto['id'] == codigo:
                print(f"Producto: {producto['nombre']}, Cantidad: {producto['cantidad']}, Precio Venta: {producto['precio_venta']}")
                return
        print(f'Producto con código {codigo} no encontrado.')
    else:
        for producto in productos:
            print(f"Producto: {producto['nombre']}, Cantidad: {producto['cantidad']}, Precio Venta: {producto['precio_venta']}")


def vender_producto(codigo, cantidad):
    for producto in productos:
        if producto['id'] == codigo:
            if producto['cantidad'] >= cantidad:
                producto['cantidad'] -= cantidad
                ventas.append({'id_producto': codigo, 'cantidad': cantidad, 'total_venta': cantidad * producto['precio_venta']})
                print(f'Venta realizada. Se vendieron {cantidad} unidades de {producto["nombre"]}.')
            else:
                print('Cantidad insuficiente en inventario.')
            return
    print(f'Producto con código {codigo} no encontrado.')


def comprar_producto(codigo, cantidad):
    for producto in productos:
        if producto['id'] == codigo:
            producto['cantidad'] += cantidad
            print(f'Se ha agregado {cantidad} unidades de {producto["nombre"]}.')
            return
    print(f'Producto con código {codigo} no encontrado.')


def resumen_ventas():
    total = 0
    for venta in ventas:
        producto = next((p for p in productos if p['id'] == venta['id_producto']), None)
        if producto:
            print(f"Producto: {producto['nombre']}, Cantidad vendida: {venta['cantidad']}, Total venta: {venta['total_venta']}")
            total += venta['total_venta']
    print(f'Total generado por ventas: {total}')


def menu_principal():
    print("Bienvenido al sistema de inventario 'Shaman Tools'")
    nombre = input("Ingrese su nombre de usuario: ")
    contraseña = input("Ingrese su contraseña: ")

    rol = iniciar_sesion(nombre, contraseña)
    if rol == 'admin':
        menu_admin()
    elif rol == 'vendedor':
        menu_vendedor()
    else:
        print("Credenciales incorrectas. Inténtelo de nuevo.")
        return


def menu_admin():
    while True:
        print("\nMenú Admin:")
        print("1. Registrar usuario")
        print("2. Agregar/Actualizar producto")
        print("3. Consultar inventario")
        print("4. Comprar productos")
        print("5. Ver resumen de ventas")
        print("6. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == '1':
            nombre = input("Nombre del nuevo usuario: ")
            contraseña = input("Contraseña: ")
            rol = input("Rol (admin/vendedor): ")
            registrar_usuario(nombre, contraseña, rol)
        elif opcion == '2':
            id_producto = int(input("ID del producto: "))
            nombre = input("Nombre del producto: ")
            precio_compra = float(input("Precio de compra: "))
            precio_venta = float(input("Precio de venta: "))
            cantidad = int(input("Cantidad disponible: "))
            unidad = input("Unidad: ")
            agregar_actualizar_producto(id_producto, nombre, precio_compra, precio_venta, cantidad, unidad)
        elif opcion == '3':
            consultar_inventario()
        elif opcion == '4':
            codigo = int(input("Código del producto: "))
            cantidad = int(input("Cantidad a comprar: "))
            comprar_producto(codigo, cantidad)
        elif opcion == '5':
            resumen_ventas()
        elif opcion == '6':
            print("Saliendo del sistema...")
            break
        else:
            print("Opción no válida. Inténtelo de nuevo.")

def menu_vendedor():
    while True:
        print("\nMenú Vendedor:")
        print("1. Consultar inventario")
        print("2. Vender producto")
        print("3. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == '1':
            consultar_inventario()
        elif opcion == '2':
            codigo = int(input("Código del producto: "))
            cantidad = int(input("Cantidad a vender: "))
            vender_producto(codigo, cantidad)
        elif opcion == '3':
            print("Saliendo del sistema...")
            break
        else:
            print("Opción no válida. Inténtelo de nuevo.")

menu_principal()
```
